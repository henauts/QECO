```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import time
import bacteria_numba_no_saturation as bacnosat
import bacteria_numba as bacsat
import sys
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
data = np.loadtxt('Results/final_populations_q_and_x_L.txt')
```

```python
qs = np.arange(0, 2.025, 0.025)
x_Ls = np.arange(0.1, 9.9, 0.05)
```

```python
X, Y = np.meshgrid(qs, x_Ls)
fig, ax = plt.subplots(figsize=(8,6))
heat = plt.contourf(X, Y, data.T[1:-2,:], cmap = 'inferno', levels = 50, vmax = 1, vmin = 0)
cbar = plt.colorbar(heat, ticks = [0.1*i for i in range(11)])
cbar.ax.set_ylabel(r'Final population $\left( N_{\mathrm{tot}} \right)$', fontsize = 14, labelpad = 10)
cbar.ax.set_yticklabels([f"{0.1*i:.1f}" for i in range(11)], fontsize = 12)
plt.xlabel(r'Substance Injection Rate $[\mu M/h]$', fontsize = 14)
plt.xticks(fontsize = 12)
plt.ylabel('Injection Position', fontsize = 14)
plt.ylim(0.1, 9.85)
plt.yticks([0.1, 5, 9.85], labels = ['Right wall', 'Center', 'Left wall'], fontsize = 12)
plt.title('Bacterial survival depending on substance injection', fontsize = 20, pad = 16)
fig.patch.set_alpha(0.2)
plt.savefig('Plots/Substance_injection_rate_position_poster.png', dpi = 800, bbox_inches = 'tight')
# plt.savefig('Plots/Substance_injection_rate_position.svg', dpi = 300, bbox_inches = 'tight')
plt.show()
```
